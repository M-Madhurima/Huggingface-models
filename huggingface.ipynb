```python
from datasets import load_dataset
```

```python
from transformers import pipeline
```

```python
from evaluate import evaluator
```

```python
eval = evaluator("text-classification")
```

```python
data = load_dataset("imdb", split="test").shuffle(seed=42).select(range(100))
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# model 1
# model = AutoModelForSequenceClassification.from_pretrained("ykacer/bert-base-cased-imdb-sequence-classification")
```

```python
# from scipy.optimize.zeros import results_c
pipe = pipeline("text-classification", model = "ykacer/bert-base-cased-imdb-sequence-classification") 
results = eval.compute(model_or_pipeline = pipe, data = data, 
                       label_mapping = {"LABEL_0":0, "LABEL_1":1},)
print(results)
```
